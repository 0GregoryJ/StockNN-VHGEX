```python
import os
import datetime

import IPython
import IPython.display
import matplotlib as mpl
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import csv
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
csv = pd.read_csv(
    "VHGEX-data.csv",
    delimiter=",")

#dataframe 5 columns, inputs; out target output is close (closing price for this (or next...) week)
df = csv[["WeekNumber", "GMEAN_OCHL"]]

GM_hist = df['GMEAN_OCHL']
GM_hist.plot()
```

```python
#create augmented matrix: X1...x5 augmented by y1,y2, where x and y are past and future GMEANS respectively
def df_to_X_y(df, windowsize=5):
    df_in_np = df.to_numpy()
    X = []
    y = []
    for i in range(len(df_in_np)-windowsize):
        row = [[a] for a in df_in_np[i:i+5]]
        X.append(row)
        label = df_in_np[i+5]
        y.append(label)
    return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X,y = df_to_X_y(GM_hist, WINDOW_SIZE)
X.shape, y.shape
```

```python
#view hist matrix
X
```

```python
#view target augment
y
```

```python
#now we can begin splitting our data into training and validation, and testing
X_train, y_train = X[:700], y[:700]
X_val, y_val = X[700:1000], y[700:1000]
X_test, y_test = X[1000:], y[1000:]

X_train.shape, y_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

#build model

model1 = Sequential()
model1.add(InputLayer((5,1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()
```

```python
#specify training callbacks
cp = ModelCheckpoint('model1/', save_best_only =True)
#compile and fit model
model1.compile(loss=MeanSquaredError(), optimizer =Adam(learning_rate = 0.001), metrics = [RootMeanSquaredError()])
```

```python
model1.fit(X_train, y_train, validation_data=(X_val,y_val), epochs = 50, callbacks=[cp])
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('model1/')
```

```python
train_predictions = model1.predict(X[600:]).flatten()
train_results = pd.DataFrame(data ={'Train Predictions':train_predictions, 'Actuals':y[600:]})
train_results
```

```python
plt.plot(train_results['Train Predictions'])
plt.plot(train_results['Actuals'])
```
